### Informações sobre os atributos

Segundo o dicionário de variáveis, os atributos estão classificados em:

* `age` - idade em anos
* `sex` - sexo do paciente
  * 0: mulher
  * 1: homem
* `cp` - tipo da dor torácica
  * 1: angina (tipo de dor no peito causada pela redução do fluxo sanguíneo para o coração) típica
  * 2: angina atípica
  * 3: dor não cardíaca
  * 4: assintomática (sem sintomas)
* `trestbps` - pressão arterial em repouso medida em mm Hg (milímetros de mércurio) no momento de admissão no hospital (considerada alta caso seja superior a 140 mm Hg)
* `chol` - colesterol sérico medido em mg/dl (sérico é um termo usado para se referir a quantidade de uma determinada substância no sangue)
* `fbs` - açúcar no sangue em jejum > 120mg/dl
  * 0: False
  * 1: True
* `restecg` - resultado do eletrocardiografia de repouso (eletrocardiografia é o processo de produção de um eletrocardiograma, gráfico de voltagem/tempo, da atividade elétrica do coração)
  * 0: normal
  * 1: anormalidades de ST-T (inversão da onda T e elevação ou depressão de > 0.05mV)
  * 2: hipertrofia ventricular esquerda provável ou definitiva (pelos critérios de Romhilt-Estes)
* `thalach` - frequência cardíaca máxima atingida, medida em batimentos por minuto (bpm)
* `exang` - angina (tipo de dor no peito causada pela redução do fluxo sanguíneo para o coração) induzida pelo exercício
  * 0: não
  * 1: sim
* `oldpeak` - depressão do segmento ST induzida pelo exercício em relação ao repouso ( segmento ST corresponde ao intervalo entre o fim da despolarização e o início da repolarização ventricular)
* `slope` - inclinação do segmento ST no pico do exercício
  * 1: ascendente
  * 2: plana
  * 3: descendente
* `ca` - número de vasos principais coloridos por fluoroscopia
* `thal` - teste de stress com tálio (teste cardiológico que mede a capacidade do coração de responder ao estresse externo em um ambiente clínico controlado, feito por estimulação farmacológica intravenosa)
  * 3: normal
  * 6: defeito fixo
  * 7: defeito reversível
* `num` - diagnóstico de doença cardíaca (diagnóstico angiográfico)
  * 0: não há doença cardíaca ( < 50% de estreitamento do diâmetro)
  * 1, 2, 3 ou 4: há doença cardíaca ( > 50% de estreitamento do diâmetro)

**Obs.:** informações entre parênteses `()` foram adicionadas para facilitar a compreensão das informações nas variáveis. Qualquer erro ou omissão é de inteira responsabilidade do autor.


### Importando os dados

Como já houve um pré-processamento dos dados, a importação do *dataset* será feito a partir do meu diretório no [Github](https://github.com/marcelorandolfo/data-science/blob/master/data/heart_disease.csv).


```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# url dos datasets
data = 'https://raw.githubusercontent.com/marcelorandolfo/data-science/master/data/heart_disease.csv'

# importar dataset da UCI
df = pd.read_csv(data)

# visualizar as primeiras 5 entradas
df.head()
```

* **0** - não há doença cardíaca

```
# variável num antes
df['num'].value_counts()
```

Com exceção da variável `num`, todas as demais são do tipo `float`. Mas se olharmos o significado das variáveis, temos dois grupos distintos: variáveis númericas e variáveis categóricas. 

Por exemplo, a variável `trestbps` é númerica, ela pode assumir qualquer valor, já que mede a pressão arterial do paciente. Mas a variável `fbs` só assume dois valores, *0* se o açúcar no sangue é menor que 120mg/dl  e *1* caso contrário. 

Nesse caso, as variáveis categóricas, como a variável `fbs`, tem como objetivo classificar o paciente dentro de um determinado grupo, ou seja, se o paciente é do grupo de homens ou mulheres, ou do grupo que tem muito ou pouco açúcar no sangue.

É importante destacar essa diferença entre as variáveis pois algumas análises estatísticas, como o boxplot por exemplo, não faz muito sentido para variáveis categóricas. Para esse tipo de variável existem outros tipos de análise que acabam sendo mais eficazes na mensagem que se quer passar.

Com isso, as variáveis do `DataFrame` podem ser divididas da seguintes forma:

* Numéricas: `age`, `trestbps`, `chol`, `thalach`, `oldpeak` e `ca`

Além disso, os pacientes que possuem valor ausente no atributo `ca` não são os mesmos que apresentam valor ausente no atributo `thal`.

Entre os dois pacientes que não possuem o resultado do teste de stress cardíaco por tálio, um foi diagnosticado com doença cardiovascular.

Quando nos deparamos com valores ausentes em um *dataset* a primeira dúvida que aparece é: como lidar com eles? Podemos substituir o valor utilizando a média, a mediana, tentar modelar esse atributo e estimar um valor possível ou simplesmente apagar essa entrada do `DataFrame`.

No caso, optarei pela última opção. Normalmente não se exclui uma entrada inteira por causa de um valor ausente, pois ao excluir essa entrada você está perdendo informação. Mas para o `DataFrame` em questão serão perdidas poucas informações.

```
# porcentagem de entradas perdidas

print('Porcentagem de entradas perdidas:\t {0:.2f}%'.format(((df['ca'].isnull().sum() + df['thal'].isnull().sum())/ df.shape[0])*100))
```

Excluindo as observações sobre os seis pacientes do `DataFrame` iremos perder aproximadamente 2% das informações. Esse valor é baixo. Caso fosse maior, por exemplo 30% do `DataFrame`, poderíamos analisar outra opção de tratamento de valores ausentes.

A questão é: qual o motivo de excluir as observações ao invés de tratá-las?

Porque as informações são sensíveis, são informações médicas. Por exemplo, para o atributo `thal` temos o resultado do teste de stress cardíaco por tálio, essa característica é muito particular do paciente. É um tanto perigoso assumir o resultado que uma pessoa apresentaria para um exame médico.

Somando os dois fatos, de que são poucas as entradas que apresentam valores ausentes e de que informações médicas são muito particulares de cada paciente, acredito que a melhor opção é excluir esses valores do `DataFrame`.

```
# criando uma cópia do dataframe
df_clean = df.copy()

# excluindo os valores ausentes
df_clean.dropna(subset = ['ca'], inplace = True)
df_clean.dropna(subset = ['thal'], inplace = True)

# tamanho do novo dataframe
print('Número de entradas/pacientes:\t {}'.format(df_clean.shape[0]))
print('Número de variáveis/atributos pessoais e médicos:\t {}'.format(df_clean.shape[1]))
```

## Análise Exploratória

A análise explatória será dividida em três: uma só para as variáveis numéricas, outra para variáveis categóricas e uma análise entre as duas categorias. A razão disso é tornar a análise mais adequada para cada tipo de variável. Por exemplo, um histograma não faz muito sentido para a variável `sex`, mas um gráfico de *waffle* sim. Para a variável `age` um gráfico de barras não passaria uma informação clara sobre a variável, um boxplot nesse caso seria melhor.

```
# dataset de variáveis numéricas
df_numer = df_clean[['age', 'trestbps', 'chol', 'thalach', 'oldpeak' ,'ca']]

# dataset de variáveis categóricas
df_cat = df_clean[['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal', 'num']]
```

### Variáveis numéricas

O primeiro passo ao analisar um `DataFrame` numérico, é visualizar as suas principais estatísticas: média, mediana, primeiro e terceiro quartil, valor mínimo e máximo e o desvio padrão.

```
df_numer.describe()
```

Observando as estatísticas do `DataFrame` de variáveis númericas, podemos perceber as seguintes características:

* Com exceção do variável `ca`, a média das variáveis é muito proxima da mediana. Isso indica que a variável está centralizada na média, ou seja, metade dos valores é menor que a média e a outra metade é maior que a média.
* Apesar da proximidade da média e mediana, temos para a variável `chol`, relativa ao colesterol do paciente, apresenta um indício de ter *outliers*. *Outlier* é um valor muito discrepante dos demais valores para o atributo, e podemos ver seu indício ao comparar o valor do terceiro quartil e o valor máximo. Podemos observar que 75% dos pacientes apresentava um colesterol menor ou igual a 276. Mas o paciente com o maior nível de colesterol no sangue apresentava um valor de 564, mais que o dobro do valor do terceiro quartil.
* A presença de *outliers* pode ser observada na variável `oldpeak` também. 75% dos pacientes apresentavam uma depressão do segmento ST menor que 1.6 milímetros, mas algum paciente apresenta um valor igual a 6.2 para o mesmo atributo.
* Na variável `ca`,  temos que metade dos pacientes não apresentaram vasos principais coloridos por fluoroscopia. Enquanto isso, para a outra metade dos pacientes esse valor oscilou de 1 até 3 vasos coloridos por fluoroscopia.

Uma outra forma de entender a distribuição dos valores das variáveis é pelo histograma. Com o histograma é possível observar a frequência que determinado valor aparece e se há ou não *outliers*.


```
# importando biblioteca seaborn
import seaborn as sns
sns.set()

# plotando os histogramas

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(16,8))

# histograma age
df_numer['age'].plot(kind="hist", ax=ax[0,0], color='r')
ax[0,0].set_title('Histograma da Idade', fontsize=14)
ax[0,0].set_ylabel('Quantidade de ocorrências')
ax[0,0].set_xlabel('Idade em anos', fontsize = 12)

# histograma trestbps
df_numer['trestbps'].plot(kind="hist", ax=ax[0,1], color='r')
ax[0,1].set_title('Histograma da Pressão Arterial', fontsize=14)
ax[0,1].set_ylabel('Quantidade de ocorrências')
ax[0,1].set_xlabel('Medida em mm/Hg', fontsize = 12)

# histograma chol
df_numer['chol'].plot(kind="hist", ax=ax[0,2], color='r')
ax[0,2].set_title('Histograma do Colesterol', fontsize=14)
ax[0,2].set_ylabel('Quantidade de ocorrências')
ax[0,2].set_xlabel('Medido em mg/dl', fontsize = 12)

# histograma thalach
df_numer['thalach'].plot(kind="hist", ax=ax[1,0], color='r')
ax[1,0].set_title('Histograma da Frequência Cardíaca',fontsize=14)
ax[1,0].set_ylabel('Quantidade de ocorrências')
ax[1,0].set_xlabel('Medido em bpm', fontsize = 12)

# histograma oldpeak
df_numer['oldpeak'].plot(kind="hist", ax=ax[1,1], color='r')
ax[1,1].set_title('Histograma da Depressão do Segmento ST', fontsize=14)
ax[1,1].set_ylabel('Quantidade de ocorrências')
ax[1,1].set_xlabel('Medido em mm', fontsize = 12)

# histograma ca
df_numer['ca'].plot(kind="hist", ax=ax[1,2], color='r')
ax[1,2].set_title('Histograma do número de vasos color. por fluoroscopia', fontsize=14)
ax[1,2].set_ylabel('Quantidade de ocorrências')
ax[1,2].set_xlabel('Quantidade de vasos', fontsize = 12)

fig.tight_layout();

```

Analisando os histogramas das variáveis numéricas do `DataFrame`, podemos observar algumas características que já observamos na análise estatística descritiva, além de outras que só é possível observar visualmente. Como por exemplo:

* Presença de valores muito altos e discrepantes para as variáveis `chol`, colesterol, e `oldpeak`, depressão do segmento ST. No caso, o *outlier* é apontado pela pequena barra à direita das demais barras. Essa barra à direita nos diz que há **poucos** pacientes com aquele nível de colesterol ou aquele tamanho de depressão do segmento ST. Ao ponto que há **muitos** pacientes com níveis menores de colesterol ou com tamanhos menores de depressão do ST.
* Além disso, pelo histograma da variável de frequência cardíaca que sua distribuição é um pouco direcionada à direita, ou seja, mais pacientes apresentam um nível de batiamento cardíaco mais elevado, enquanto menos pacientes apresentam um nível de batimento mais baixo. Também é possível observar a possibilidade um *outlier* à esquerda.
* De forma contrária à variável de frequência cardíaca, a variável de pressão arterial é mais direcionada à esquerda. Muitos pacientes apresentam uma pressão arterial mais baixo, enquanto menos pacientes apresentam uma pressão arterial mais elevada.
* Para a variável de idade, podemos observar a presença de pacientes um pouco mais velhos, o que faz sentido, já que estamos tratando de pacientes que estavam em um hospital.
* Pelo histograma da variável de vasos coloridos por fluoroscopia, podemos visualizar o que já era esperado. A maioria dos pacientes não apresenta nenhum vaso principal colorido no exame.

Outra análise gráfica importante de ser feita é por meio do boxplot. Com ele, é possível visualizar o intervalo entre quartis e a presença, ou não, de *outliers*. Isso porque o boxplot segue a seguinte estrutura:

<center><img alt="bp" width="60%" src="https://drive.google.com/uc?id=1-5O80O8L3GXNvQD7bI3RmO7EHbn84sJ_"></center>

Nesse caso, todo *ponto* em um boxplot representa que há um valor na variável que pode ser considerado, por definição, como um *outlier*.

```
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(16,8))

# histograma age
df_numer.boxplot(['age'], ax=ax[0,0], vert=False, widths=0.6, grid=False)
ax[0,0].set_yticks([])
ax[0,0].set_xlabel('Anos', fontsize = 12)
ax[0,0].set_title('Boxplot da Idade', fontsize=14)

# histograma trestbps
df_numer.boxplot(['trestbps'], ax=ax[0,1], vert=False, widths=0.6, grid=False)
ax[0,1].set_yticks([])
ax[0,1].set_xlabel('mm/Hg', fontsize = 12)
ax[0,1].set_title('Boxplot da Pressão Arterial', fontsize=14)

# histograma chol
df_numer.boxplot(['chol'], ax=ax[0,2], vert=False, widths=0.6, grid=False)
ax[0,2].set_yticks([])
ax[0,2].set_xlabel('mg/dl', fontsize = 12)
ax[0,2].set_title('Boxplot do Colesterol', fontsize=14)

# histograma thalach
df_numer.boxplot(['thalach'], ax=ax[1,0], vert=False, widths=0.6, grid=False)
ax[1,0].set_yticks([])
ax[1,0].set_xlabel('bpm', fontsize = 12)
ax[1,0].set_title('Boxplot da Frequência Cardíaca', fontsize=14)

# histograma oldpeak 
df_numer.boxplot(['oldpeak'], ax=ax[1,1], vert=False, widths=0.6, grid=False)
ax[1,1].set_yticks([])
ax[1,1].set_xlabel('mm', fontsize = 12)
ax[1,1].set_title('Boxplot da Depressão do Segmento ST', fontsize=14)

# histograma ca 
df_numer.boxplot(['ca'], ax=ax[1,2], vert=False, widths=0.6, grid=False)
ax[1,2].set_yticks([])
ax[1,2].set_xlabel('Quantidade', fontsize = 12)
ax[1,2].set_title('Boxplot do número de vasos color. por fluoroscopia', fontsize=14)

fig.tight_layout();
```

Pelos boxplot das variáveis `chol` e `oldpeak`, confirmamos nossa hipótese de presença de *outliers* à direita. Além disso, confirmamos também que há um *outlier* para a variável de frequência cardíaca, como sugerimos na análise do histograma. 

Na análise do histograma, vimos que a variável de pressão arterial era mais direcionada à esquerda. Na ocasião não sugerimos a presença de *outliers*, mas de acordo com o boxplot, há sim *outliers* à direita. O mesmo pode ser observado para a variável do número de vasos coloridos por fluoroscopia.

Para a variável de idade, não há a presença de *outliers*. Nesse caso, temos que a variável é bem distribuída, sem valores discrepantes.

Considerando a variável idade, podemos analisar se pacientes mais velhos apresentam um nível de colesterol mais alto ou baixo, ou se a frequência cardíaca fica mais lenta ou mais rápida. Nesse caso, podemos utilizar o gráfico de dispersão para encontrar essas relações.

```
# plotar o gráfico de dispersão age x demais variáveis
fig = plt.figure(figsize=(18,10))
fig.subplots_adjust(hspace=0.3, wspace=0.3)

# age x trestbps
ax1 = plt.subplot(221)
sns.regplot(y='age', x='trestbps', scatter_kws={'s': 60.0, 'color': 'firebrick'},
            line_kws={'color': 'gray'}, data=df_clean, ax=ax1)
ax1.set_title('Idade x Pressão Arterial', fontsize = 16)
ax1.set_ylabel('Anos')
ax1.set_xlabel('mm/Hg')

# age x chol
ax2 = plt.subplot(222)
sns.regplot(y='age', x='chol', scatter_kws={'s': 60.0, 'color': 'firebrick'},
            line_kws={'color': 'gray'}, data=df_clean, ax=ax2)
ax2.set_title('Idade x Colesterol', fontsize = 16)
ax2.set_ylabel('Anos')
ax2.set_xlabel('mg/dl')

# age x thalach
ax3 = plt.subplot(234)
sns.regplot(y='age', x='thalach', scatter_kws={'s': 60.0, 'color': 'firebrick'},
            line_kws={'color': 'gray'}, data=df_clean, ax=ax3)
ax3.set_title('Idade x Frequência Cardíaca', fontsize = 16)
ax3.set_ylabel('Anos')
ax3.set_xlabel('bpm')

# age x oldpeak
ax4 = plt.subplot(235)
sns.regplot(y='age', x='oldpeak', scatter_kws={'s': 60.0, 'color': 'firebrick'},
            line_kws={'color': 'gray'}, data=df_clean, ax=ax4)
ax4.set_title('Idade x Depressão do Segmento ST', fontsize = 16)
ax4.set_ylabel('Anos')
ax4.set_xlabel('mm')

# age x ca
ax5 = plt.subplot(236)
sns.regplot(y='age', x='ca', scatter_kws={'s': 60.0, 'color': 'firebrick'},
            line_kws={'color': 'gray'}, data=df_clean, ax=ax5)
ax5.set_title('Idade x Número de vasos color. por fluoroscopia', fontsize = 16)
ax5.set_ylabel('Anos')
ax5.set_xlabel('Quantidade')

fig.show()
```

No gráfico de dispersão, temos também uma reta de regressão. Não é o objetivo deste *notebook* explicar as características de uma reta de regressão, ela só esta sendo usada como suporte na análise de relação entre as variáveis. No caso, se essa reta é ascendente, indica que as duas variáveis análisadas possuem relação positiva entre si, enquanto se a reta é descendente, a relação entre as variáveis é negativa. Logo, podemos observar que:

### Variáveis categóricas

Para as variáveis categóricas do `DataFrame`, as análises serão feitas com base na proporção dos valores. Por exemplo, para a variável sexo dos pacientes, temos que para pacientes mulheres o valor indicado é `0`, e pacientes homens o valor indicado é `1`. Não faz sentido analisar um boxplot nesse caso, porque os valores só podem ser `0` ou `1`, mas um gráfico de *waffle* que nos indica a proporção de homens e mulheres no `DataFrame` é muito útil. 

Além do gráfico de *waffle*, existem outros gráficos adequados a variáveis categóricas, como por exemplo o gráfico de barras e o gráfico de pizza. O gráfico de barras é útil, mas em alguns casos a proporção em que uma barra é maior que a outra não é tão simples de se identificar. Por sua vez, a utilização de gráficos de pizza é bastante criticada. Isso ocorre porque só é possível ter uma noção clara da diferença entre a proporção de valores quando essa diferença é muito elevada. Além disso, gráficos de pizza para variáveis que possuem mais de duas categorias, são bastante confusos.

Por isso, a análise será feita utilizando gráficos de *waffle*, que visualmente são mais úteis para apresentar proporções, e também possuem uma apresentação mais bonita. Caso seja do interesse saber mais sobre esse tipo de gráfico, a sua documentação pode ser vista neste [link](https://github.com/gyli/PyWaffle).

#### Análise de correlação

Quando dizemos que duas variáveis são correlacionadas entre si, queremos dizer que ambas possuem uma relação. No caso, essa relação pode ser positiva ou negativa. 

Em uma correlação positiva temos que quando o valor de uma variável aumenta, o valor de uma outra variável também aumenta. Na correlação negativa temos que quando o valor de uma variável aumenta, o valor de uma outra variável diminui.

O valor do índice de correlação vai de -1 até 1, onde -1 a correlação entre as variáveis é perfeitamente negativa, e o valor 1 indica que as variáveis são correlacionadas de forma perfeita e positiva. Além disso, correlações que apresentam valor menor que 0.39, ou -0.39, nos diz que as variáveis são correlacionadas de forma fraca. Logo, só iremos considerar as correlações que forem maiores que 0.40, ou -0.40.

```
# heatmap (mapa de calor)
fig, ax = plt.subplots(figsize=(16,8))

corr = df_clean.corr().round(2)
sns.heatmap(corr, cmap='RdBu', linecolor='white', annot=True, ax=ax, vmin=-1.00, vmax= 1.00)
ax.set_title('Histograma da Idade', fontsize=14)
```

De acordo com o *heatmap*, podemos observar as seguintes correlações:

* `cp` e `num` apresentam uma correlação positiva.
* `thalach` e `num` apresentam uma correlação negativa.
* `exang` e `num` apresentam uma correlação positiva.
* `slope` e `oldpeak` apresentam uma correlação positiva.
* `oldpeak` e `num` apresentam uma correlação positiva.
* `ca` e `num` apresentam uma correlação positiva.
* `thal` e `num` apresentam uma correlação positiva.

Se observarmos atentamente, veremos que em todas as situações de correlação citadas há pelo menos uma variável categórica. Isso acaba dificultando um pouco a análise de correlação.

Por exemplo, a variável `thal` é correlacionada de forma positiva com `num`. Nesse caso poderiamos pensar que aumentando `thal`, `num` também aumenta. Mas ambas são variáveis categóricas, não é possível aumentar a variável `num` até um valor maior que `1` ser apresentado, independentemente do que aconteça com a variável `thal`. 

Existem outros métodos de análise de correlação para variáveis categóricas, mas que não serão abordadas nesse *notebook*. O que manteremos é só a percepção de que a principal variável do nosso `DataFrame`, `num`, é correlacionada com as demais variáveis. Logo, temos que ser diagnosticado com doença cardiovascular é correlacionado com o estado de saúde do paciente.
